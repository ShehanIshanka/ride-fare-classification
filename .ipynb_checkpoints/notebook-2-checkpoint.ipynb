```python
# Haversine distance
from math import radians, cos, sin, asin, sqrt

def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 6371 # Radius of earth in kilometers. Use 3956 for miles
    return c * r
```

```python
training_raw = pd.read_csv('train.csv').drop(["tripid"],axis=1).dropna()
testing_raw = pd.read_csv('test.csv')
```

```python
training_raw["label"][training_raw["label"]=="correct"] = 1
training_raw["label"][training_raw["label"]=="incorrect"] = 0
```

```python
training_raw.head()
```

### New Feature addition

```python
training_dataset =  training_raw.dropna()
training_dataset['label'] = training_dataset.label.astype('int')
testing_dataset =  testing_raw
```

```python
def add_new_features(df):
    df["actual_trip_time"] = df["duration"] - df["meter_waiting"]
    df["actual_trip_fare"] = df["fare"] - df["meter_waiting_fare"]
    df["haversine_distance"] = df.apply(lambda x : haversine(x.pick_lon,x.pick_lat,x.drop_lon,x.drop_lat)*1000,axis=1)
    
    df["trip_fare_per_second"] =  df.apply(lambda x : 0 if x.duration==0 else x.fare/x.duration,axis=1)
    df["trip_fare_per_distance"] =  df.apply(lambda x : 0 if x.haversine_distance==0 else x.fare/x.haversine_distance,axis=1)
    df["average_vehicle_speed"] = df.apply(lambda x : 0 if x.actual_trip_time==0 else x.haversine_distance/x.duration,axis=1)
    
    df["actual_trip_fare_per_second"] =  df.apply(lambda x : 0 if x.actual_trip_time==0 else x.actual_trip_fare/x.actual_trip_time,axis=1)
    df["meter_waiting_fare_per_second"] =  df.apply(lambda x : 0 if x.meter_waiting==0 else x.meter_waiting_fare/x.meter_waiting,axis=1)
    df["actual_trip_fare_per_distance"] =  df.apply(lambda x : 0 if x.haversine_distance==0 else x.actual_trip_fare/x.haversine_distance,axis=1)
    df["average_vehicle_speed_without_waiting"] = df.apply(lambda x : 0 if x.actual_trip_time==0 else x.haversine_distance/x.actual_trip_time,axis=1)
    
add_new_features(testing_dataset)    
add_new_features(training_dataset)  
```

### Model Training with new features

```python
columns = ['actual_trip_fare_per_second',
           'average_vehicle_speed',
           'average_vehicle_speed_without_waiting']
```

```python
training_dataset[columns+['label']].head()
```

```python
final_testing_dataset =  testing_dataset[columns]
X = training_dataset[columns].values
Y = training_dataset['label'].values
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,random_state=42) 
```

### Logistic Regression

```python
reg = linear_model.LogisticRegression() 
reg.fit(X_train, y_train) 

print("TRAIN SET")
y_pred_train = reg.predict(X_train) 
print("Logistic Regression model accuracy(in %):",  metrics.accuracy_score(y_train, y_pred_train)*100)
scores = precision_recall_fscore_support(y_train,y_pred_train,average='micro')
print("Precison : "+str(scores[0])+", Recall : "+str(scores[1])+", F Score : "+str(scores[2]))

print("TEST SET")
y_pred = reg.predict(X_test) 
print("Logistic Regression model accuracy(in %):",  metrics.accuracy_score(y_test, y_pred)*100)
scores = precision_recall_fscore_support(y_test,y_pred,average='micro')
print("Precison : "+str(scores[0])+", Recall : "+str(scores[1])+", F Score : "+str(scores[2]))
```

```python
submission = pd.DataFrame()
submission['tripid'] = testing_raw['tripid']
submission["Prediction"]= reg.predict(final_testing_dataset.values)
```

```python
submission.to_csv("submissions/submission_with_new_features_logreg.csv",index=False)
```
